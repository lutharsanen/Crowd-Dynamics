```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.exp(2j * np.pi * np.arange(8) / 8)
b =  np.fft.fft(np.exp(2j * np.pi * np.arange(8) / 8))
```

```python
x = [i for i in range(8)]

fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_size_inches(20, 5)

ax1.plot(x,a)
ax2.plot(x,b)

plt.show()
```

```python
sinus_value = [ np.sin(i) for i in range(50)]
fft_sinus = np.fft.fft(sinus_value)
x_value = [i for i in range(50)]
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_size_inches(20, 5)

ax1.plot(x_value,sinus_value)
ax2.plot(x_value,fft_sinus)

plt.show()
```
